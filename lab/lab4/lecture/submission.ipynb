```python
import plotly.express as px
import pandas as pd


baby = pd.read_csv('data/baby.csv')
baby.head(2)


import plotly.graph_objects as go
fig = go.Figure(data=go.Violin(x= baby['Maternal Smoker'], y= baby['Birth Weight'], box_visible=True,
                               line_color='black',
                               meanline_visible=True, fillcolor='lightseagreen', opacity=0.6,
                               ))
fig.update_layout(title='Distribution of Birth Weight by Maternal Smoker Status',xaxis_title='Maternal Smoker',yaxis_title='Birth Weight')
fig.update_layout(yaxis_zeroline=True, xaxis_showgrid=True, yaxis_showgrid=True)
fig.show()
```
